```python
import pandas as pd
import warnings
warnings.filterwarnings(action='ignore')

from rdkit import Chem
from rdkit.DataStructs.cDataStructs import TanimotoSimilarity
```

```python
smifile = "../Data/Indole_-_cytoprotekcja_NEW_improved_smiles.xlsx"
data = pd.read_excel(smifile)
data = data['SMILES'][:45]

initial_structures_new = list(set(data))
print(len(initial_structures_new))
```

```python
tested = ['O=C(NC(C(F)=C1)=O)N1CC2=CNC3=CC=CC=C32', 'S=C(NC(C=C1C2=CC=CC=C2)=O)N1CC3=CNC4=CC=CC=C43', 'S=C(NC(C(CCC)=C1)=O)N1CC2=CNC3=CC=CC=C32', 'S=C(NC(C=C1CCC)=O)N1CC2=CNC3=CC=CC=C32']
tested_ = pd.DataFrame(data=tested, columns=['tested_smiles'])
#tested__ = tested.copy()
```

```python
for element in initial_structures_new:
    tested_[element] = 0
```

```python
new_file = tested_.reset_index()
new_file = new_file.drop(columns=['index'])
new_file
```

```python
for element in initial_structures_new:
    finger_1 = Chem.RDKFingerprint(Chem.MolFromSmiles(element))
    for i, smi in enumerate(new_file['tested_smiles']):
        finger_2 = Chem.RDKFingerprint(Chem.MolFromSmiles(smi))
        ta_sim = TanimotoSimilarity(finger_1, finger_2)
        new_file[element].loc[i] = ta_sim
        if ta_sim == 1:
            new_file[element][i] = 100
```
